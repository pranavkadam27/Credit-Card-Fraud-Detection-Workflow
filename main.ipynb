# Problem Statement
This project aims to detect fraudulent credit card transactions using anonymized transaction features (V1â€“V28) and transaction amount (Amount). We explore statistical relationships between features and fraud occurrence, test hypotheses regarding feature importance and transaction value, and validate findings through machine learning models (logistic regression, decision tree).

## Hypothesis 1:
- Hâ‚€: The mean transaction amount for fraud = mean for non-fraud
- Hâ‚: The means differ

Performing Z test

```python
%pip install -qU statsmodels
```

## Î± = 0.05

```python
import numpy as np

fraud = df[df['Class']==1]['Amount']
nonfraud = df[df['Class']==0]['Amount']

mean1, mean2 = np.mean(fraud), np.mean(nonfraud)
std1, std2 = np.std(fraud, ddof=1), np.std(nonfraud, ddof=1)
n1, n2 = len(fraud), len(nonfraud)

z_value = (mean1 - mean2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))
print(f"Z = {z_value:.3f}")

z_critical = 1.96
if abs(z_value) > z_critical:
    print("Reject H0: Significant difference in means.")
else:
    print("Fail to reject H0: No significant difference.")
```

With a 5% level of significance, we fail to reject H0, which indicates that the mean transaction amount of fraud = mean transaction amount of non-fraud

## Hypothesis - 2 (Chi square test of independence):
- Hâ‚€: Fraud occurrence is independent of transaction amount category.
- Hâ‚: Fraud occurrence depends on transaction amount category.

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2

df['Amount_bin'] = pd.cut(df['Amount'], bins=5, labels=False)
contingency = pd.crosstab(df['Amount_bin'], df['Class'])
observed = contingency.values
expected = np.outer(contingency.sum(axis=1), contingency.sum(axis=0)) / contingency.values.sum()

chi2_stat = ((observed - expected)**2 / expected).sum()
dof = (observed.shape[0]-1) * (observed.shape[1]-1)

chi2_critical = chi2.ppf(0.95, dof)
print(f"ChiÂ² = {chi2_stat:.3f}, Ï‡Â² critical = {chi2_critical:.3f}")

if chi2_stat > chi2_critical:
    print("Reject H0: Fraud depends on transaction amount bin.")
else:
    print("Fail to reject H0.")
```

With a 5% level of significance, we fail to reject the H0.

## Hypothesis 3 (Correlation Test):
- Hâ‚€: No correlation between feature V12 and fraud label.
- Hâ‚: There is correlation.

```python
from scipy.stats import pearsonr, t
import numpy as np

corr, _ = pearsonr(df['V12'], df['Class'])
n = len(df)

t_value = corr * np.sqrt((n - 2) / (1 - corr**2))

t_critical = t.ppf(1 - 0.025, n - 2)

print(f"Correlation r = {corr:.3f}, t = {t_value:.3f}, t_critical = Â±{t_critical:.3f}")

if abs(t_value) > t_critical:
    print("Reject Hâ‚€: Correlation is significant.")
else:
    print("Fail to reject Hâ‚€: No significant correlation.")
```

## Hypothesis 4 (F-Test for Equality of Variances):

- Hâ‚€: Ïƒâ‚Â² = Ïƒâ‚‚Â²
- Hâ‚: Ïƒâ‚Â² â‰  Ïƒâ‚‚Â²

```python
from scipy.stats import f

x1 = df[df['Class']==1]['V7']
x2 = df[df['Class']==0]['V7']
f_stat = np.var(x1, ddof=1) / np.var(x2, ddof=1)
d1, d2 = len(x1)-1, len(x2)-1

f_critical_low = f.ppf(0.025, d1, d2)
f_critical_high = f.ppf(0.975, d1, d2)
print(f"F = {f_stat:.3f}, Critical range: ({f_critical_low:.3f}, {f_critical_high:.3f})")

if f_stat < f_critical_low or f_stat > f_critical_high:
    print("Reject H0: Variances differ.")
else:
    print("Fail to reject H0.")
```

## 1. BASELINE MODEL - Random Forest on Full Dataset

**Problem Statement:** Credit Card Fraud Detection  
**Model:** Random Forest Classifier  
**Dataset Size:** 568,630 samples with 30 features

### Running Time Analysis:
- **Training Time Complexity:** O(n Ã— m Ã— k Ã— log(n))
  - n = number of samples (568,630)
  - m = number of features (30)
  - k = number of trees (100)
  - log(n) = depth of each tree

- **Prediction Time Complexity:** O(k Ã— log(n))
  - k = number of trees
  - log(n) = average tree depth


```python
# Prepare the full dataset
X_full = X.copy()
y_full = y.copy()

print(f"Full Dataset Shape: {X_full.shape}")
print(f"Number of samples: {X_full.shape[0]}")
print(f"Number of features: {X_full.shape[1]}")

# Split the data
X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(
    X_full, y_full, test_size=0.2, random_state=42, stratify=y_full
)

print(f"\nTraining samples: {X_train_full.shape[0]}")
print(f"Test samples: {X_test_full.shape[0]}")
```

```python
# Train Random Forest on FULL dataset and measure time
print("\n" + "="*60)
print("TRAINING ON FULL DATASET")
print("="*60)

start_time = time.time()
rf_full = RandomForestClassifier(
    n_estimators=100, 
    max_depth=15, 
    random_state=42,
    n_jobs=-1
)
rf_full.fit(X_train_full, y_train_full)
train_time_full = time.time() - start_time

# Predictions
start_pred = time.time()
y_pred_train_full = rf_full.predict(X_train_full)
y_pred_test_full = rf_full.predict(X_test_full)
pred_time_full = time.time() - start_pred

# Calculate metrics
train_acc_full = accuracy_score(y_train_full, y_pred_train_full)
test_acc_full = accuracy_score(y_test_full, y_pred_test_full)
precision_full = precision_score(y_test_full, y_pred_test_full)
recall_full = recall_score(y_test_full, y_pred_test_full)
f1_full = f1_score(y_test_full, y_pred_test_full)

print(f"\nâœ“ Training Time: {train_time_full:.2f} seconds")
print(f"âœ“ Prediction Time: {pred_time_full:.2f} seconds")
print(f"\nðŸ“Š Model Performance:")
print(f"   Train Accuracy: {train_acc_full:.4f} ({train_acc_full*100:.2f}%)")
print(f"   Test Accuracy:  {test_acc_full:.4f} ({test_acc_full*100:.2f}%)")
print(f"   Precision:      {precision_full:.4f}")
print(f"   Recall:         {recall_full:.4f}")
print(f"   F1-Score:       {f1_full:.4f}")
```

## 2. RANDOMIZED SCALING TECHNIQUES

We will apply three randomized scaling techniques:

### A. Random Sampling (Uniform Random Sampling)
- **Method:** Randomly sample a subset of training data
- **Reason:** Reduces dataset size while maintaining statistical properties
- **Complexity:** O(1) per sample selection
- **Sampled Size:** 20% of original data (~90k samples)

### B. Random Projection (Johnson-Lindenstrauss Transform)
- **Method:** Project high-dimensional data to lower dimensions using random matrices
- **Reason:** Reduces feature dimensionality while approximately preserving pairwise distances
- **Complexity:** O(d Ã— k) where d=original dims, k=projected dims
- **Target Dimensions:** 15 features (50% reduction)

### C. Combined Method (Sampling + Projection)
- **Method:** Apply both random sampling and random projection
- **Reason:** Maximum computational savings in both sample and feature space
- **Benefits:** Faster training, lower memory footprint


### Method 1: Random Sampling


```python
print("\n" + "="*60)
print("METHOD 1: RANDOM SAMPLING")
print("="*60)

# Random sampling: Take 20% of the training data
sample_size = int(0.2 * len(X_train_full))
sample_indices = np.random.choice(len(X_train_full), size=sample_size, replace=False)

X_train_sampled = X_train_full.iloc[sample_indices]
y_train_sampled = y_train_full.iloc[sample_indices]

print(f"\nOriginal training size: {len(X_train_full)}")
print(f"Sampled training size: {len(X_train_sampled)}")
print(f"Reduction: {(1 - sample_size/len(X_train_full))*100:.1f}%")

# Train model on sampled data
start_time = time.time()
rf_sampled = RandomForestClassifier(
    n_estimators=100, 
    max_depth=15, 
    random_state=42,
    n_jobs=-1
)
rf_sampled.fit(X_train_sampled, y_train_sampled)
train_time_sampled = time.time() - start_time

# Predictions on full test set
y_pred_test_sampled = rf_sampled.predict(X_test_full)

# Calculate metrics
test_acc_sampled = accuracy_score(y_test_full, y_pred_test_sampled)
precision_sampled = precision_score(y_test_full, y_pred_test_sampled)
recall_sampled = recall_score(y_test_full, y_pred_test_sampled)
f1_sampled = f1_score(y_test_full, y_pred_test_sampled)

print(f"\nâœ“ Training Time: {train_time_sampled:.2f} seconds")
print(f"âœ“ Speedup: {train_time_full/train_time_sampled:.2f}x faster")
print(f"\nðŸ“Š Model Performance:")
print(f"   Test Accuracy:  {test_acc_sampled:.4f} ({test_acc_sampled*100:.2f}%)")
print(f"   Precision:      {precision_sampled:.4f}")
print(f"   Recall:         {recall_sampled:.4f}")
print(f"   F1-Score:       {f1_sampled:.4f}")
print(f"\nðŸ“‰ Accuracy Drop: {(test_acc_full - test_acc_sampled)*100:.2f}%")
```

### Method 2: Random Projection (Johnson-Lindenstrauss)


```python
print("\n" + "="*60)
print("METHOD 2: RANDOM PROJECTION")
print("="*60)

# Random Projection: Reduce dimensions from 30 to 15
n_components = 15
print(f"\nOriginal features: {X_train_full.shape[1]}")
print(f"Projected features: {n_components}")
print(f"Dimensionality reduction: {(1 - n_components/X_train_full.shape[1])*100:.1f}%")

# Fit random projection
start_projection = time.time()
rp = GaussianRandomProjection(n_components=n_components, random_state=42)
X_train_projected = rp.fit_transform(X_train_full)
X_test_projected = rp.transform(X_test_full)
projection_time = time.time() - start_projection

print(f"\nâœ“ Projection Time: {projection_time:.2f} seconds")

# Train model on projected data
start_time = time.time()
rf_projected = RandomForestClassifier(
    n_estimators=100, 
    max_depth=15, 
    random_state=42,
    n_jobs=-1
)
rf_projected.fit(X_train_projected, y_train_full)
train_time_projected = time.time() - start_time

# Predictions
y_pred_test_projected = rf_projected.predict(X_test_projected)

# Calculate metrics
test_acc_projected = accuracy_score(y_test_full, y_pred_test_projected)
precision_projected = precision_score(y_test_full, y_pred_test_projected)
recall_projected = recall_score(y_test_full, y_pred_test_projected)
f1_projected = f1_score(y_test_full, y_pred_test_projected)

total_time_projected = projection_time + train_time_projected

print(f"âœ“ Training Time: {train_time_projected:.2f} seconds")
print(f"âœ“ Total Time (projection + training): {total_time_projected:.2f} seconds")
print(f"âœ“ Speedup: {train_time_full/total_time_projected:.2f}x faster")
print(f"\nðŸ“Š Model Performance:")
print(f"   Test Accuracy:  {test_acc_projected:.4f} ({test_acc_projected*100:.2f}%)")
print(f"   Precision:      {precision_projected:.4f}")
print(f"   Recall:         {recall_projected:.4f}")
print(f"   F1-Score:       {f1_projected:.4f}")
print(f"\nðŸ“‰ Accuracy Drop: {(test_acc_full - test_acc_projected)*100:.2f}%")
```

### Method 3: Combined (Random Sampling + Random Projection)


```python
print("\n" + "="*60)
print("METHOD 3: COMBINED (SAMPLING + PROJECTION)")
print("="*60)

# Apply both techniques
# First: Random Sampling (20%)
sample_size_combined = int(0.2 * len(X_train_full))
sample_indices_combined = np.random.choice(len(X_train_full), size=sample_size_combined, replace=False)
X_train_sampled_combined = X_train_full.iloc[sample_indices_combined]
y_train_sampled_combined = y_train_full.iloc[sample_indices_combined]

print(f"\nSampling: {len(X_train_full)} â†’ {len(X_train_sampled_combined)} samples")

# Second: Random Projection (15 dims)
start_projection_combined = time.time()
rp_combined = GaussianRandomProjection(n_components=n_components, random_state=42)
X_train_combined = rp_combined.fit_transform(X_train_sampled_combined)
X_test_combined = rp_combined.transform(X_test_full)
projection_time_combined = time.time() - start_projection_combined

print(f"Projection: {X_train_full.shape[1]} â†’ {n_components} features")
print(f"\nâœ“ Projection Time: {projection_time_combined:.2f} seconds")

# Train model
start_time = time.time()
rf_combined = RandomForestClassifier(
    n_estimators=100, 
    max_depth=15, 
    random_state=42,
    n_jobs=-1
)
rf_combined.fit(X_train_combined, y_train_sampled_combined)
train_time_combined = time.time() - start_time

# Predictions
y_pred_test_combined = rf_combined.predict(X_test_combined)

# Calculate metrics
test_acc_combined = accuracy_score(y_test_full, y_pred_test_combined)
precision_combined = precision_score(y_test_full, y_pred_test_combined)
recall_combined = recall_score(y_test_full, y_pred_test_combined)
f1_combined = f1_score(y_test_full, y_pred_test_combined)

total_time_combined = projection_time_combined + train_time_combined

print(f"âœ“ Training Time: {train_time_combined:.2f} seconds")
print(f"âœ“ Total Time (projection + training): {total_time_combined:.2f} seconds")
print(f"âœ“ Speedup: {train_time_full/total_time_combined:.2f}x faster")
print(f"\nðŸ“Š Model Performance:")
print(f"   Test Accuracy:  {test_acc_combined:.4f} ({test_acc_combined*100:.2f}%)")
print(f"   Precision:      {precision_combined:.4f}")
print(f"   Recall:         {recall_combined:.4f}")
print(f"   F1-Score:       {f1_combined:.4f}")
print(f"\nðŸ“‰ Accuracy Drop: {(test_acc_full - test_acc_combined)*100:.2f}%")
```

## 4. ASYMPTOTIC COMPLEXITY ANALYSIS

### Baseline Model (Random Forest on Full Dataset)

**Training Complexity:**
- O(n Ã— m Ã— k Ã— log(n))
  - n = 454,904 (training samples)
  - m = 30 (features)
  - k = 100 (trees)
  - Dominant term: O(n Ã— m Ã— k Ã— log n)

**Prediction Complexity:**
- O(k Ã— log(n))
  - Each tree: O(log n)
  - k trees total

---

### Method 1: Random Sampling

**Sampling Complexity:** O(s) where s = sample size

**Training Complexity:**
- O(s Ã— m Ã— k Ã— log(s))
  - s = 0.2n â‰ˆ 90,980 samples
  - Improvement: **5x faster** in theory (since s = n/5)

**Space Complexity:**
- Reduced from O(n Ã— m) to O(s Ã— m) = **80% memory savings**

---

### Method 2: Random Projection

**Projection Complexity:**
- O(n Ã— m Ã— d) where d = target dimensions (15)
- One-time cost for transforming data

**Training Complexity:**
- O(n Ã— d Ã— k Ã— log(n))
  - d = 15 (vs m = 30)
  - Improvement: **2x faster** per tree construction

**Space Complexity:**
- Feature space reduced from O(m) to O(d) = **50% reduction**

---

### Method 3: Combined (Sampling + Projection)

**Total Complexity:**
- Sampling: O(s)
- Projection: O(s Ã— m Ã— d)
- Training: O(s Ã— d Ã— k Ã— log(s))

**Overall Improvement:**
- **Sample reduction:** 5x from sampling
- **Feature reduction:** 2x from projection
- **Combined speedup:** ~10x theoretical improvement

**Space Complexity:**
- O(s Ã— d) = O(0.2n Ã— 0.5m) = **90% memory reduction**
